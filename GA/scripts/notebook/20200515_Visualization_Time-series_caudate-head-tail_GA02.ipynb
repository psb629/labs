## make SPMG2_rwdtm2_trained.pkl & SPMG2_rwdtm2_untrained.pkl from their preprocessed data

```python
def getrwdtm2(subj, run):
    data_dir = glm_dir + subj
    rwdtm2 = []
    censor = []
    xmat = data_dir + '/X.xmat.%s.r%02d.SPMG2.1D' % (subj, run)
    censor_list = preproc_dir + subj + '/preprocessed/motion_%s.r%02d_censor.1D' % (subj, run)
    
    # get the list of censored/uncensored TRs
    with open(censor_list) as f:
        lines_censor = f.readlines()who
        
        for censor_val in lines_censor:
            censor.append(int(censor_val))
    
    assert len(censor) == 1096
    
    # get the xmat data containing rwdtm#2
    with open(xmat) as f:
        lines_xmat = f.readlines()
    
    xmat = lines_xmat[23:-2]
    
    # get the full rwdtm#2 data, treating values for censored TRs as nan
    count_xmat = 0
    for idx, censor_val in enumerate(censor): # 1096
        if censor_val == 1: # uncensored:
            rwdtm2_val = float(xmat[count_xmat].split(' ')[8])
            rwdtm2.append(rwdtm2_val)
            count_xmat += 1
        elif censor_val == 0: # censored:
            rwdtm2.append('nan')
        else:
            print('There is something wrong!\n')
    
    assert len(rwdtm2) == 1096
    
    return rwdtm2
```

```python
subj_list = subj_list_GD
rwdtm2_trained = {}
rwdtm2_untrained = {}

for subj in subj_list:
    rwdtm2_trained_temp = []
    rwdtm2_untrained_temp = []
    
    for run in range(1, 7):
        rwdtm2_temp = getrwdtm2(subj, run)
        if run < 4:
            rwdtm2_trained_temp.extend(rwdtm2_temp)
        else:
            rwdtm2_untrained_temp.extend(rwdtm2_temp)
    
    rwdtm2_trained[subj] = rwdtm2_trained_temp
    rwdtm2_untrained[subj] = rwdtm2_untrained_temp
```

```python
with open(reggress_dir+'SPMG2_rwdtm2_trained.pkl', 'wb') as handle:
    pickle.dump(rwdtm2_trained, handle)

with open(reggress_dir+'SPMG2_rwdtm2_untrained.pkl', 'wb') as handle:
    pickle.dump(rwdtm2_untrained, handle)

# with open('filename.pickle', 'rb') as handle:
#     b = pickle.load(handle)
```

```python
def cor2mni(x, y, z):
    T = np.array([(-2.683, 0, 0, 95.2465), (0, -2.683, 0, 130.686), (0, 0, 2.683, -77.2465)])
    mni = np.dot(T, np.array((x, y, z, 1)).reshape(-1, 1))
    mni = np.round(mni)
    
    return -mni[0][0], -mni[1][0], mni[2][0]
```

```python
def draw_plot(subj, ax, is_train, is_early):
    if is_train:
        runs = [1, 2, 3]
        train_str = 'Trained'
    else:
        runs = [4, 5, 6]
        train_str = 'Untrained'
        
    if not is_early:
        subj = subj.replace('A', 'B')
    
    if is_train and is_early:
        _, cdh_signal, cdt_signal, _, rwd = results[subj, 'Trained']
        _, _, loc_heads, val_heads, loc_tails, val_tails, _, _ = list(zip(*[results[subj, train_str, run] for run in runs]))
        title = 'early-trained'
    elif not is_train and is_early:
        _, cdh_signal, cdt_signal, _, rwd = results[subj, 'Untrained']
        _, _, loc_heads, val_heads, loc_tails, val_tails, _, _ = list(zip(*[results[subj, train_str, run] for run in runs]))
        title = 'early-untrained'
    elif is_train and not is_early:
        cdh_signal, _, _, cdt_signal, rwd = results[subj.replace('A', 'B'), 'Trained']
        loc_heads, val_heads, _, _, _, _, loc_tails, val_tails = list(zip(*[results[subj, train_str, run] for run in runs]))
        title = 'late-trained'
    elif not is_train and not is_early:
        cdh_signal, _, _, cdt_signal, rwd = results[subj.replace('A', 'B'), 'Untrained']
        loc_heads, val_heads, _, _, _, _, loc_tails, val_tails = list(zip(*[results[subj, train_str, run] for run in runs]))
        title = 'late-untrained'
        
    ax.set_ylabel('%s' % (title), labelpad=20, rotation=0, fontsize=20, ha='right')
    
    ax.plot(np.arange(1096*3), rwd, '-', label='success rate', linewidth=3, color='#AAAAAA')
    ax.plot(np.arange(1096*3), cdh_signal, '-' , label='caudate head', linewidth=1, color='#000000')
    ax.plot(np.arange(1096*3), cdt_signal, '--', label='caudate tail', linewidth=1, color='#FF0000')
    ax.plot((1096, 1096), (-7, 7), 'k-', alpha=0.2)
    ax.plot((1096*2, 1096*2), (-7, 7), 'k-', alpha=0.2)
    ax.set_xlim(0, 1096*3)
    ax.set_ylim(-5, 5)
    
    ax.set_xticks([0, 1096, 1096*2])
    ax.set_xticklabels(['        Run %d' % r for r in runs], fontsize=15)
        
    df = pd.DataFrame(np.array([cdh_signal, cdt_signal, rwd.reshape(-1)]).T, 
                      columns=['head', 'tail', 'success rate'])

    head_corr = df.corr(method='pearson')['head']['success rate']
    tail_corr = df.corr(method='pearson')['tail']['success rate']

    ax.text(25, 4,'caudate head corr R = %.2f' % (head_corr), va='center', style='italic', fontsize=15)
    ax.text(25, -4, 'caudate tail corr R = %.2f' % (tail_corr), va='center', style='italic', fontsize=15)

#     ax.text(1096-300, 4, '%d, %d, %d (T = %.2f)' % (*cor2mni(*loc_heads[0]), val_heads[0]), va='center', fontsize=15)
#     ax.text(1096-300, -4, '%d, %d, %d (T = %.2f)' % (*cor2mni(*loc_tails[0]), val_tails[0]), va='center', fontsize=15)

#     ax.text(1096*2-300, 4, '%d, %d, %d (T = %.2f)' % (*cor2mni(*loc_heads[1]), val_heads[1]), va='center', fontsize=15)
#     ax.text(1096*2-300, -4, '%d, %d, %d (T = %.2f)' % (*cor2mni(*loc_tails[1]), val_tails[1]), va='center', fontsize=15)

#     ax.text(1096*3-300, 4, '%d, %d, %d (T = %.2f)' % (*cor2mni(*loc_heads[2]), val_heads[2]), va='center', fontsize=15)
#     ax.text(1096*3-300, -4, '%d, %d, %d (T = %.2f)' % (*cor2mni(*loc_tails[2]), val_tails[2]), va='center', fontsize=15)
    
    ax.legend(loc='center left', bbox_to_anchor=(1.0, 1.0))
```

```python
fig, ax = plt.subplots(4, 1, figsize=(30, 10))

subj = 'GA02'

# early, trained

draw_plot(subj, ax[0], is_train=True, is_early=True)

# early, untrained

draw_plot(subj, ax[1], is_train=False, is_early=True)

# late, trained

draw_plot(subj, ax[2], is_train=True, is_early=False)

# late, untrained

draw_plot(subj, ax[3], is_train=False, is_early=False)
        
fig.tight_layout()
```
