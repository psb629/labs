```python
import pandas as pd
import scipy.stats
import pickle
import numpy as np
```

# Localizer

```python
df = pd.read_csv("/Volumes/clmnlab/GA/MVPA/LSS_pb02_MO_short_duration/roi_mvpa/20200313/localizer/20200313_localizer_dec-acc.csv")
df.tail()
```

```python
roi_names = sorted(set(df['roi_name']))

lines = []

for roi_name in roi_names:
    for stage in ['Early', 'Late']:
        for mapping in ['Practiced', 'Unpracticed']:
            accuracies = df[(df['Mapping'] == mapping) & (df['Stage'] == stage) & (df['roi_name'] == roi_name)]['Mean Accuracy']
            res = scipy.stats.ttest_1samp(a=accuracies, popmean=0.25)
            lines.append((roi_name, stage, mapping, res.statistic, res.pvalue))
```

```python
result_df = pd.DataFrame(lines, columns=['cluster_name', 'stage', 'mapping', 't-value', 'p-value'])
result_df
```

```python
result_df.to_csv('20200327_MVPA_localizer_statistics.csv')
```

```python
subj_list = [
    'GA01', 'GA02', 'GA05', 'GA07', 'GA08', 'GA11', 'GA12', 'GA13', 'GA14', 'GA15',
    'GA18', 'GA19', 'GA20', 'GA21', 'GA23', 'GA26', 'GA27', 'GA28', 'GA29', 'GA30',
    'GA31', 'GA32', 'GA33', 'GA34', 'GA35', 'GA36', 'GA37', 'GA38', 'GA42', 'GA44'
]
```

```python
lines = []

for roi_name in roi_names:
    early_accuracies = [float(df[(df['Stage'] == 'Early') & (df['Mapping'] == 'Practiced') & (df['roi_name'] == roi_name) & (df['subj'] == subj)]['Mean Accuracy']) for subj in subj_list]
    late_accuracies = [float(df[(df['Stage'] == 'Late') & (df['Mapping'] == 'Practiced') & (df['roi_name'] == roi_name) & (df['subj'] == subj)]['Mean Accuracy']) for subj in subj_list]
    res = scipy.stats.ttest_rel(early_accuracies, late_accuracies)
    lines.append((roi_name, res.statistic, res.pvalue))
```

```python
result_df = pd.DataFrame(lines, columns=['cluster_name', 't-value', 'p-value'])
result_df
```

```python
result_df.to_csv('20200327_MVPA_localizer_paired_statistics.csv')
```
